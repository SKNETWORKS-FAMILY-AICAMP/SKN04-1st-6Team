##### 현대 FAQ

**1. 정적 크롤링(request) 선택의 문제**
> 문제 : 질문 드롭다운 & 카테고리 리스트 클릭과 같은 사용자 동작이 필요하여 
> 정적 크롤링으로 처리할 수 없었음
>
> 해결 : 동적 크롤링(selenium)을 사용

**2. 화면 크기에 따라 pagenation 번호 상이**
> 문제 : 화면 접근 시 page번호가 화면 크기에 따라서 번호가 다르게 나타남
>
> 해결 : 화면 크기를 1980*1080으로 해상도 고정하여 화면 크기를 표준화함

**3. 홈페이지 내 태그 요소 인식 불가**
> 문제 : 특정 요소를 클릭하려고 할 때, 해당 요소가 존재하지 않거나 
> 인식되지 않는 문제 발생
>
> 해결 : 'WebDriverWait'를 활용하여 실제 존재하고 클릭이 가능할 때까지 대기하도록 구현

**4. 마지막 페이지 도착 후 상단 카테고리 선택 불가**
> 문제 : 마지막 페이지 도착 후 다음 카테고리 선택이 안되는 문제 발생,
> 'click()'를 사용하였지만 화면에 가려져 보이지 않아서 작동되지 않음
>
> 해결 : .click() 함수 대신 javascript 버튼 동작으로 강제 클릭되도록 수행
